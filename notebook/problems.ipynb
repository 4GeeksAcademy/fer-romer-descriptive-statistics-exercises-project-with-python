# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Also known as Gaussian distribution
# Generate an array of 100 elements following a normal distribution (mean = 0 and standard deviation = 1)
normal_array = np.random.normal(size = 100) 

# Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom                         
chi_square_array = np.random.chisquare( 
                                        df=3,   # --> degrees of freedom (tell you how many numbers in your grid are actually independent)
                                            size=100    # --> number of elements
)
      
#print(f"Normal vector: {normal_array}")
#print(f"Chi-square vector: {chi_square_array}")
````

The Chi-Square Distribution, X^2, is the result of summing up 'df' random independent variables from the Standard Normal Distribution.

For 'df'=3, the corresponding Ch-Square looks like: (look below: subplot(2))

The values are on average a lot greater and more left skewed. This makes sense as we are more likely to sample a high value number when we have more degrees of freedom. The summation also naturally leads to higher values.

## Statistical clarification of arrays

````python
# Figure with subplots to show both distributions in parallel (figsize = set the size of the figure)
plt.figure(figsize=(12, 6))  


# Chart 1: Normal Distribution
plt.subplot(1, 2, 1)  # Create the first subplot (1 row, 2 columns, chart 1)

    # Histogram with density line (KDE)
sns.histplot(normal_array,
              bins=20,
                kde=True,   # (bool): If True, compute a kernel density estimate to smooth the distribution and show on the plot as (one or more) line(s). Only relevant with univariate data.
                  alpha=0.7,
                    color='darkblue',
                      edgecolor='black')  

plt.title("Distribucin Normal",
          fontstyle = 'oblique')  # Chart Title   

plt.xlabel("(bin)Intervalos ( , ]",
           fontstyle = 'italic')  # X axis label

plt.ylabel("Frecuencia Absoluta (ni)",
           fontstyle = 'italic')  # Y axis label

plt.grid(True)



# Chart 2: Chi-Square Distribution
plt.subplot(1, 2, 2)  # Create the second subplot (1 row, 2 columns, chart 2)

    # Histogram with density line (KDE)
sns.histplot(chi_square_array,
              bins=20, 
                kde=True,  # (bool): If True, compute a kernel density estimate to smooth the distribution and show on the plot as (one or more) line(s). Only relevant with univariate data.
                  alpha=0.7,
                    color='red',
                      edgecolor='black')  

plt.title("Distribucin Chi-cuadrado (3 grados de libertad)",
          fontstyle = 'oblique')  # Chart Title

plt.xlabel("(bin)Intervalos ( , ]",
           fontstyle = 'italic')  # X axis label

plt.ylabel("Frecuencia Absoluta (ni)",
           fontstyle = 'italic')  # Y axis label

plt.grid(True)



plt.tight_layout()  # Adjust margins to avoid overlapping
plt.show()  
````

# -------Clarification-------

Descriptive statistics consists of analyzing a "sample" using a series of parameters 

In this case we are going to work with two continuous samples ("Yes" TAKE DECIMALS) grouped in the form of intervals (Approximation).

- X axis: Each box (bin) of the histogram is an interval of the values (we have indicated that bin=20 so we group the values into 20 intervals).

- Y axis: Absolute frequency (ni), which is the number of elements in the sample that meet a certain characteristic --> The sum of all the "ni" gives the total number of elements (in this case n=100).

___

Both graphs are histograms with a Kernel Density Estimation (KDE) that smoothes the curve, allowing the density of the data to be better seen.
___

Normal Distribution: It has a symmetrical shape, with most of the values concentrated around the mean (0).

Chi-square Distribution: It is asymmetric and skewed to the right, with more values in the lower ranges and a long tail towards the higher values.
___

## Measures of central tendency

#### Mean

Arithmetic mean: (sum of values) / total number of elements
- The sample mean is not very resistant to the presence of outliers (it can seriously affect the final value).

#### Median

It is the value that occupies the central position of the sample (sorted).
- It is resistant to the presence of outliers

#### Mode

It is the value that is most repeated

````python
import statistics as stats


# Normal Array

mean_normal_array = stats.mean(normal_array) # Calculate the Mean
median_normal_array = stats.median(normal_array) # Calculate the Median
mode_normal_array = stats.mode(normal_array) # Calculate the Mode


# Chi-Square Array

mean_chi_square_array = np.mean(chi_square_array) # Calculate the Mean
median_chi_square_array = np.median(chi_square_array) # Calculate the Median
mode_chi_square_array = stats.mode(chi_square_array) # Calculate the Mode

# Create a chart whit 2 subplots 
fig, axs = plt.subplots(1, 2, figsize=(12, 6))


# Normal Distribution

# We use a histogram to visualize the frequency of values in the normal distribution, with a density curve added.
sns.histplot(normal_array,
              bins=20,
                kde=True,
                  ax=axs[0],
                    color='skyblue')

# Vertical lines for the mean, median and mode
axs[0].axvline(mean_normal_array,
                color='red',
                  linestyle='--',
                    label=f'Mean: {mean_normal_array:.3f}')

axs[0].axvline(median_normal_array,
                color='green',
                  linestyle='-',
                    label=f'Median: {median_normal_array:.3f}')

axs[0].axvline(mode_normal_array,
                color='blue',
                  linestyle='-.',
                    label=f'Mode: {mode_normal_array:.3f}')

axs[0].set_title("Distribucin Normal",
          fontstyle = 'oblique')  # Chart Title

axs[0].set_xlabel("(bin)Intervalos ( , ]",
           fontstyle = 'italic')  # X axis label

axs[0].set_ylabel("Frecuencia Absoluta (ni)",
           fontstyle = 'italic')  # Y axis label

axs[0].legend()


# Chi-Square Distribution

# We use a histogram to visualize the frequency of values in the chi-square distribution, with a density curve added.
sns.histplot(chi_square_array,
              bins=20,
                kde=True,
                  ax=axs[1],
                    color='lightcoral')

# Vertical lines for the mean, median and mode
axs[1].axvline(mean_chi_square_array,
                color='red',
                  linestyle='--',
                    label=f'Mean: {mean_chi_square_array:.3f}')

axs[1].axvline(median_chi_square_array,
                color='green', linestyle='-',
                  label=f'Median: {median_chi_square_array:.3f}')

axs[1].axvline(mode_chi_square_array,
                color='blue', linestyle='-.',
                  label=f'Mode: {mode_chi_square_array:.3f}')

axs[1].set_title("Distribucin Chi-cuadrado (3 grados de libertad)",
          fontstyle = 'oblique')  # Chart Title

axs[1].set_xlabel("(bin)Intervalos ( , ]",
           fontstyle = 'italic')  # X axis label

axs[1].set_ylabel("Frecuencia Absoluta (ni)",
           fontstyle = 'italic')  # Y axis label

axs[1].legend()

# Adjust margins to avoid overlapping
plt.tight_layout()

plt.show()
````

## Measures of dispersion
 (Are numerical values that describe how varied the data are in set)

Range

````python
# The difference between the maximun value and the minimun value of a data

range_normal = max(normal_array) - min(normal_array)
range_chi = max(chi_square_array) - min(chi_square_array)

print(f"Range Normal vector: {range_normal}")
print(f"Range chi_square vector: {range_chi}")
````

Variance and standard deviation

````python
'''Both metrics measure the same thing. They indicate how far, on average, the values are from the mean.
    However, the standart deviation is a measure used to be able to work with initial units of measurement,
        while the variance, although at first sight it may seen an unnecessary calculation, is calculated
             to be able to obtain other parameters'''

variance_normal = stats.variance(normal_array)
variance_chi_square = stats.variance(chi_square_array)

std_normal = stats.stdev(normal_array)
std_chi_square = stats.stdev(chi_square_array)

print(f"Normal variance: {variance_normal} and std: {std_normal}")
print(f"Chi variance: {variance_chi_square} and std: {std_chi_square}")
````

## Shape measures

Skewness

````python
# Measures the degrees of symmetry of a distribution
from scipy.stats import skew

skew_normal = skew(normal_array)
skew_chi_square = skew(chi_square_array)

print(f"Normal skewness: {skew_normal}")
print(f"Chi skewness: {skew_chi_square}")
````

Kurtosis

````python
# Is the fourth central moment divided by the square of the variance

from scipy.stats import kurtosis

kurtosis_normal = kurtosis(normal_array)
kurtosis_chi_square = kurtosis(chi_square_array)

print(f"Normal kurtosis: {kurtosis_normal}")
print(f"Chi kurtosis: {kurtosis_chi_square}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO

import math

data = [4, 2, 5, 8, 6]

# Calculate the standart deviation of the data

def st_dev(data):
   
   n = len(data) # Cantidad de nmeros que hay

   if n <= 1:
      return 0.0 # Si hay un solo elemento, no tiene sentido calcular la desviacin estndar
   
   mean = sum(data) / n # La media de los nmeros

   diff_cuadrados = [(_ - mean) ** 2 for _ in data] # Suma de todos sus resultados individuales, menos la puntuacin media, elevando todos al cuadrado

   variance = sum(diff_cuadrados) / (n-1) # Obtenemos la varianza

   return math.sqrt(variance)

print("Data: ")
print(data)
print('\n')
print(f"Standard Deviation: {st_dev(data)}")
````
